Imorting importatnt libraries.


```
import seaborn as sns
import pandas as pd
```

Reading the csv files

```
odi=pd.read_csv("/content/CricinfoICC ODI Championship.csv")
odi
```

```
wodi=pd.read_csv("/content/CricinfoICC Women's ODI Team Rankings.csv")
wodi
```

Trimming the team names

```
def trim(x):
  a=x[:3]
  return a
```

```
odi['team']=odi['team'].apply(trim)
odi
```

```
wodi['team']=wodi['team'].apply(trim)
wodi
```

Increasing the fontsize

```
import matplotlib as mpb
mpb.rcParams['font.size']=25
```

Plotting the barplot for team vs points




```
import matplotlib.pyplot as plt
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('team','points',data=odi)
```

Plotting the barplot for team vs rating


```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('team','rating',data=odi)
```

Plotting the horizontal barplot for rating vs team.


```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('rating','team',data=odi)
```

Plotting the horizontal barplot for rating vs team for women ODI data


```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('rating','team',data=wodi)
```

Plotting the horizontal barplot for matches vs team


```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('matches','team',data=wodi)
```

Plotting the horizontal barplot for team vs points and segregating them further on the basis of gender


```
wodi['gender']='Female'
```

```
odi['gender']='Male'
odi
```

```
comb=pd.concat([wodi,odi],ignore_index=True)
comb
```

```
f,ax=plt.subplots(figsize=(20,10))

ax=sns.barplot('team','points',hue='gender',data=comb)
```

Plotting the horizontal scatterplot for team vs matches and segregating them further on the basis of gender


```
f,ax=plt.subplots(figsize=(20,10))
ax=sns.scatterplot('team','matches',hue='gender',data=comb)
```
